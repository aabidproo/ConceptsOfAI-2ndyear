```python
import pandas as pd

medical_data = pd.read_csv("Dataset/medical_students_dataset.csv")

df = pd.DataFrame(medical_data)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#height data were not in meter so converting to meter
df["Height_m"] = df["Height"]/100

mising_bmi_mask = df["BMI"].isnull()

# BMI = weight(kg) / [height(m)]Â²

df.loc[mising_bmi_mask, "BMI"] = df.loc[mising_bmi_mask, "Weight"] / (df.loc[mising_bmi_mask, "Height_m"]**2)

df.drop("Height_m", axis=1, inplace=True)

mising_bmi_mask.sum()

```

```python
df.isnull().sum()
```

```python
print(f"\nNumber of duplicate rows: {df.duplicated().sum()}")
```

```python
df.shape
```

```python
if df.duplicated().sum() > 0:
    print(f"Duplicate rows found ")
    print(df[df.duplicated(keep=False)])

    df = df.drop_duplicates(keep="first")
    print(f"after removing shape: {df.shape}")
else:
    print("No duplicate rows found Dataset is clean")
```

```python
df.isnull().sum()
```

```python


df["height_m"] = df['Height']/100

can_fix_bmi = df["BMI"].isnull() & df["Height"].notnull() & df["Weight"].notnull()


df.loc[can_fix_bmi, "BMI"] = df.loc[can_fix_bmi, "Weight"]/(df.loc[can_fix_bmi, "height_m"]**2)

df.drop('height_m', axis=1, inplace=True )

df["BMI"].isnull().sum()
```

```python
df.isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns.tolist()

categorical_cols = df.select_dtypes(exclude="float64").columns.tolist()

numeric_cols
categorical_cols
```

```python


for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        most_common = df[col].mode()[0]
        df[col].fillna(most_common, inplace=True)
        

df[categorical_cols].isnull().sum()
```

```python
df[numeric_cols].isnull().sum()
```

```python
from sklearn.impute import KNNImputer

body_cols = ['Height', 'Weight', 'BMI']

knn_imputer = KNNImputer(n_neighbors=5)

df[body_cols] = knn_imputer.fit_transform(df[body_cols])
```

```python
df[body_cols].isnull().sum()
```

```python
for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        median_value = df[col].median()
        df[col].fillna(median_value, inplace=True)

        
```

```python
df[numeric_cols].isnull().sum()
df[categorical_cols].isnull().sum()
```
