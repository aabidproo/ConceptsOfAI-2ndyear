```python
import numpy as np

empty_array_1 = np.empty((2, 2))
print("Empty array 2x2:\n", empty_array_1)

ones_array = np.ones((4, 2))
print("\nAll-one array 4x2:\n", ones_array)

filled_array = np.full((3, 3), fill_value=7, dtype=int)
print("\nFilled array 3x3 with 7:\n", filled_array)

sample_array = np.array([[5, 6, 7], [8, 9, 10]])
zeros_like_array = np.zeros_like(sample_array)
print("\nZeros like sample_array:\n", zeros_like_array)

ones_like_array = np.ones_like(sample_array)
print("\nOnes like sample_array:\n", ones_like_array)

new_list = [1, 2, 3, 4]
array_from_list = np.array(new_list)
print("\nArray from list:\n", array_from_list)
```

```python
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

add_xy = x + y
print(add_xy)

sub_xy = x - y
print(sub_xy)

mul_x = x * 3
print(mul_x)

square_x = x ** 2
print(square_x)

dot_vw = np.dot(v, w)
dot_xv = np.dot(x, v)
dot_xy = np.dot(x, y)
print(dot_vw)
print(dot_xv)
print(dot_xy)

concat_xy_rows = np.concatenate((x, y), axis=0)
print(concat_xy_rows)

concat_vw_cols = np.column_stack((v, w))
print(concat_vw_cols)

try:
    concat_xv = np.concatenate((x, v), axis=0)
    print(concat_xv)
except Exception as e:
    print("Error:", e)
```

```python
A = np.array([[3, 4],
              [7, 8]])

B = np.array([[5, 3],
              [2, 1]])

A_inv = np.linalg.inv(A)
identity_check = A.dot(A_inv)
print(identity_check)


AB = A.dot(B)
BA = B.dot(A)

print("AB =\n", AB)
print("BA =\n", BA)

left_side = (A.dot(B)).T
right_side = B.T.dot(A.T)

print(left_side)
print(right_side)

```

```python
A_sys = np.array([[ 2, -3,  1],
                  [ 1, -1,  2],
                  [ 3,  1, -1]])

B_sys = np.array([-1, -3, 9])

A_inv_sys = np.linalg.inv(A_sys)
X_solution = A_inv_sys.dot(B_sys)

print("Solution:", X_solution)
```

```python
X_linalg = np.linalg.solve(A_sys, B_sys)
print("Solution",X_linalg)
```
